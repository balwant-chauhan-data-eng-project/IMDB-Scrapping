```python
 

# Define a regex pattern to extract movie information
pattern = r'(\d+)\.\s+(.+?)(\d{4})(\d{0,2}h\s+\d{0,2}m)?([A-Za-z\s\-]+)(\d+\.\d)Rate'

# Find movie information using the regex pattern
movie_info = re.findall(pattern, paragraph)

# Create a list to store movie details
movielist = []

# Iterate through the extracted movie_info
for i in movie_info:
    # Create a dictionary to store details of a single movie
    movie_details = {
        'ranking': i[0],           # Extract ranking from the regex match
        'movie': i[1].strip(),     # Extract movie name from the regex match and remove leading/trailing spaces
        'year_of_release': i[2],   # Extract year of release from the regex match
        'duration': re.sub(r'[UA]', '', i[3].strip()) if i[3] else None,
                                  # Extract duration from the regex match and remove 'UA' if present, set to None if not present
                                  # Extract rating from the regex match and remove leading/trailing spaces
        'rating': i[5]               # Extract rate from the regex match
    }
    
    # Append the movie details dictionary to the movielist
    movielist.append(movie_details)

# Create a DataFrame from the list of movie details
df = pd.DataFrame(movielist)

# Display the DataFrame with movie details (without index)
display(df)

 

 
```
